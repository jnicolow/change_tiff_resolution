# Batch upsample site TIFFs (bilinear + bicubic)

Runs `convert_site_tiffs.convert_sites_to_higher_resolution`, which mirrors the workflow in `change_tiff_res_test.ipynb`: discover all `.tif` / `.tiff` files under each site folder, upsample by an integer scale factor, and write outputs to **`data/sat_images/<site>/<bilinear|bicubic>/...`** (relative subfolders preserved).

**Set the working directory to this repo root** (`change_tiff_resolution`) before running, or adjust `PROJECT_ROOT` below.

```python
import sys
from pathlib import Path

try:
    PROJECT_ROOT = Path(__file__).resolve().parent
except NameError:
    PROJECT_ROOT = Path.cwd()

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from change_tiff_res.convert_site_tiffs import convert_sites_to_higher_resolution
```

```python
import os

# Parent folder that contains one subfolder per site (same layout as change_tiff_res_test.ipynb).
source_dir = Path(
    os.path.join(
        os.path.dirname(os.getcwd()),
        "SDS_performance_analysis",
        "data",
        "sat_images",
    )
)

# Folder names under source_dir (e.g. "californiatorreypines").
site_names = [
    "australianarrabeen",
    "australiastocktonbeach",
    "australiabyronbay",
    "australiamonavale",
    "californiatorreypines",
    "northcarolinaduck",
    "francetrucvert",
    "hawaiiwaikiki",
    "alaskabarterisland",
    "massachusettscapecod",
    "floridamadeirabeach"
]

upsample_scale = 2

# Default: PROJECT_ROOT / "data" / "sat_images"
output_root = PROJECT_ROOT / "data" / "sat_images"

print("source_dir:", source_dir)
print(f'{site_names=}')
print("output_root:", output_root.resolve())
```

```python
written = convert_sites_to_higher_resolution(
    source_dir,
    site_names,
    output_root=output_root,
    upsample_scale=upsample_scale,
    techniques=("bilinear", "bicubic"),
)

print(f"Wrote {len(written)} file(s).")
for p in written[:20]:
    print(p)
if len(written) > 20:
    print("...")
```
